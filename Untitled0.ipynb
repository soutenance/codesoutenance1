```
#import csv
import pubchempy 
from pubchempy import *
#from pubchempy import get_compounds, Compound
v=[]
bb=[]
k=[]
listm=[]
data=[]
sss=''

generatedMol = open("newfile.txt","w")
generatedCorpus = open("corpus","w")
#fichiercsv=csv.reader(f)

f= open('pIGC50T-pyriformis.csv')
for i in f :
  data.append(i.split(';')[1])
  
for i in data :
  v.append(i)
  list1=[]
  for s in range (len(i)-1):
      for ss in range(s+1,len(i)):
          sss=i[s:ss]
          MOL=Chem.MolFromSmiles(sss)
          bb.append(MOL)
          if bb[0]!=None :
            k.append(sss)
            list1.append(sss)
          bb=[]
  listm.append(list1)
  listm.append(v)
  generatedMol.write(str(list1))
  generatedMol.write("\n")
  generatedMol.flush()
  v=[]
print(listm)
generatedCorpus.write(str(listm))
```

```
from gensim.models import Word2Vec
model=Word2Vec(listm, min_count=1)
word=list(model.wv.vocab)
print(word)
```
